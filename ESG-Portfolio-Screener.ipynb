```python
import pandas as pd
import os

# Creating a sample dataset so the code runs immediately
# In a real setup, I'd just upload the 'esg_data.csv' file
if not os.path.exists('esg_data.csv'):
    sample_data = {
        'Ticker': ['AAPL', 'MSFT', 'TSLA', 'XOM', 'BP', 'GOOGL', 'AMZN', 'PEP', 'UNH', 'CVX'],
        'Company': ['Apple', 'Microsoft', 'Tesla', 'Exxon', 'BP Oil', 'Google', 'Amazon', 'Pepsi', 'UnitedHealth', 'Chevron'],
        'Sector': ['Tech', 'Tech', 'Auto', 'Energy', 'Energy', 'Tech', 'Consumer', 'Consumer', 'Healthcare', 'Energy'],
        'ESG_Score': [84, 88, 75, 40, 35, 80, 55, 70, 60, 30],
        'Annual_Return': [25.0, 28.0, 12.5, 8.0, 6.5, 22.0, 18.0, 10.0, 15.0, 7.2]
    }
    pd.DataFrame(sample_data).to_csv('esg_data.csv', index=False)

# Loading the market data
df = pd.read_csv('esg_data.csv')

# Simple function to filter out companies based on their ethics score
def screen_stocks(min_score):
    # Separating 'Force for Good' companies from high-risk ones
    ethical_picks = df[df['ESG_Score'] >= min_score]
    return ethical_picks

# Setting a threshold (JPMC style sustainability filter)
# Anything below 70 is flagged as a risk
min_acceptable_score = 70
good_stocks = screen_stocks(min_acceptable_score)

# Calculating basic stats for the report
market_avg = df['Annual_Return'].mean()
ethical_avg = good_stocks['Annual_Return'].mean()

print("--- JPMC SUSTAINABLE PORTFOLIO AUDIT ---")
print(f"Filtering for companies with ESG Score >= {min_acceptable_score}")
print(f"Total approved companies: {len(good_stocks)}")

print("\nApproved Ethical Portfolio:")
print(good_stocks[['Ticker', 'Company', 'ESG_Score', 'Annual_Return']])

print("\n--- PERFORMANCE SUMMARY ---")
print(f"Market Average Return: {market_avg:.2f}%")
print(f"Ethical Portfolio Return: {ethical_avg:.2f}%")

if ethical_avg > market_avg:
    print("\nResult: Ethical investing outperformed the general market.")
else:
    print("\nResult: Standard market slightly outperformed the ethical filter.")
```
